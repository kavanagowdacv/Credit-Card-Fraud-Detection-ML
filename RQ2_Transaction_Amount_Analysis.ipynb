```python
# =========================================
# RQ2 - Transaction Amount Analysis
# =========================================

# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

print("Libraries Loaded Successfully")

# Load Dataset
data = pd.read_csv(
    "/kaggle/input/datasets/kavanacv/creditcard-2023/credit_card_frauds.csv"
)

print("\nDataset Loaded Successfully")

# Display First 5 Rows
print("\nFirst 5 Rows:")
print(data.head())

# =========================================
# Research Question 2
# How does transaction amount differ
# between fraudulent and normal transactions?
# =========================================

# Create Statistics Table
rq2_table = data.groupby('is_fraud')['amt'].describe()

print("\nTransaction Amount Statistics:")
print(rq2_table)

# Save Table
rq2_table.to_csv("RQ2_Amount_Statistics.csv")

# Create Boxplot Figure
plt.figure(figsize=(6,4))

data.boxplot(column='amt', by='is_fraud')

plt.title("Transaction Amount by Fraud Class")
plt.suptitle("")

plt.xlabel("Class (0 = Normal, 1 = Fraud)")
plt.ylabel("Transaction Amount")

# Save Figure
plt.savefig("RQ2_Amount_Boxplot.pdf")

# Show Figure
plt.show()

print("\nRQ2 Completed Successfully")
```
