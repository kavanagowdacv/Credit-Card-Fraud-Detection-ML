```python
# =========================================
# RQ6 - Fraud by Transaction Category
# =========================================

# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import os

print("Libraries Loaded Successfully")

# =========================================
# FIND DATASET AUTOMATICALLY
# =========================================

csv_path = ""

for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        if filename.endswith('.csv'):
            csv_path = os.path.join(dirname, filename)

print("\nDataset Path Found:")
print(csv_path)

# =========================================
# LOAD DATASET
# =========================================

data = pd.read_csv(csv_path)

print("\nDataset Loaded Successfully")

# Display First 5 Rows
print("\nFirst 5 Rows:")
print(data.head())

# =========================================
# FRAUD CATEGORY ANALYSIS
# =========================================

fraud_data = data[data['is_fraud'] == 1]

category_counts = fraud_data['category'].value_counts()

print("\nFraud Count by Category:")
print(category_counts)

# Save Table
category_counts.to_csv(
    "RQ6_Fraud_By_Category.csv"
)

# =========================================
# BAR GRAPH
# =========================================

plt.figure(figsize=(12,6))

category_counts.plot(kind='bar')

plt.title("Fraudulent Transactions by Category")

plt.xlabel("Category")
plt.ylabel("Number of Fraud Transactions")

plt.xticks(rotation=45)

plt.tight_layout()

# Save Figure
plt.savefig("RQ6_Fraud_By_Category.pdf")

plt.show()

print("\nRQ6 Completed Successfully")
```
